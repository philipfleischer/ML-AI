## Celler

Den mest fundamentale byggesteinen for Jupyter notebooks er celler. En celle er en boks slik som denne, og finnes i to former, Markdown og Kode. Markdown er et tekstformat, som er det denne cellen bruker. Kode-celler vil kunne kjøre Python kode interaktivt, som vi vil se på straks.

Prøv å dobbeltklikk denne cellen. Da går man fra *command-mode* til *edit-mode*. Det vil åpne den i rå-tekst. Her kan dere redigere det som står. For å få den tilbake til kompilert Markdown kan dere trukke "play" symbolet øverst til høyre i cellen eller bruke hurtigtasten `shift` + `enter`.

Prøv å rediger overskriften (som begynner med `##`) i denne cellen og kjør den.

Man kan navigere seg mellom celler ved å bruke piltastene opp og ned i command-mode. Dersom man trykker `enter` vil man åpne cellen for redigering. Man kan endre rekkefølgen på celler ved å trykke `alt / option` + `pil_opp` eller `pil_ned` i command-mode (altså *ikke* mens en cellen redigeres, trykk `shift` + `enter` først).

## Litt mer om celler

Husk at dersom cellen er åpen for redigering er dere i *edit-mode*, og dersom den er kjørt med `shift` + `enter` er dere i *command-mode*. I command mode kan dere sette inn en ny celle over ved å kun klikke `a`, eller under med `b`. Dette vil gi en kode-celle som dere kan gjøre om til Markdwon ved å trykke `m`. For å slette trykk `d` + `d`. Dere kan bytte rekkefølgen på celler med `alt / option` + `pil_opp` eller `pil_ned`, eller kun bruke piltastene for å navigere.

Øverst i vinduet er det noen kommandoer, som å lage celler. Her er det også noen andre nyttige funksjoner. Knappen `Run All` vil kjøre alle kode-cellene i dokumentet. Det vil være nyttig når dere får større notebooks og trenger å kjøre gjennom alt. Knappen `Restart` vil fjerne all minnen fra cellene. Det vil si at variabler og funksjoner som er definert fra kode-celler som er kjørt vil bli slettet (ikke koden eller cellene, bare minnet fra at de har blitt kjørt). Dere kan også fjerne output med `Clear All Output`. Det er lurt å bruke `Restart` og så `Clear All Outputs` og `Run All` en gang i blant for å unngå bugs, og det er nødvendig å gjøre før dere leverer obliger.

## Noen viktige tips

Dere vil bli vandt med å bruke Jupyter notebooks i ukesoppgavene og obligene, så vi kommer ikke til å gå igjennom mye mer her, men la oss se på noen triks og noen vanlige opphav for bugs og forvirring.

Jupyter notebooks er fleksible og praktiske, men gir også opphav til noen nye bugs. Dersom dere for eksempel kjører en kode-celle som definerer en variabel `my_var`, vil den være i minne. Andre kode-celler som bruker denne variablen vil da kunne bruke `my_var`. Dette gjelder *også om kode-cellen som definerer `my_var` slettes*, som kan føre til bugs. Man vil altså ikke med det første merke at man bruker en udefinert variabel, siden den fortsatt ligger i minnet, men dersom man restarter kernelen eller kjører notebooken fra en annen datamaskin vil ikke koden kjøre.

Tilsvarende bugs kan også oppstå ved at man kjører kode i forskjellig rekkefølge eller flere ganger. Notebooks burde designet slik at de skal kjøres fra topp til bunn én gang. For å hindre slike bugs, kjør `Restart` etterfulgt av `Clear All Outputs` og `Run All` i menyen øverst. Dere må også gjøre dette før dere leverer obliger og se over at alt virker som det skal.

Selv om interaktive notebooks er praktiske, er det ikke alt de passer til. Større prosjekter blir ofte forvirrende dersom det er definert ut ifra kode i mange forskjellige celler, og det blir lettere å holde oversikt med en god filstruktur istedenfor. Det er vanlig å definere større prosjekter med vanlig filstruktur og `.py` filer og deretter bruke notebooks til å demonstrere bruken av koden ved å importere funksjonalitet fra koden og kjøre den i notebooksene med kommentarer i Markdown. I obligen deres kommer tester og hjelpefunksjoner til å ligge vedlagt som `.py` filer, slik at notebooksene blir ryddigere.
