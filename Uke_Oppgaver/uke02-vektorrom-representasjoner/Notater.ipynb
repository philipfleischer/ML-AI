# Seminar - Vektorrom-representasjoner

#### Lag din egen personvektor
Trekk:          Type:       Eksempel:
Kaffe vs. Te    binær       0 = te, 1 = kaffe
Høyde           tall(cm)    165
Antall språk    tall        5

##### Svar
Vi kan representere en person som en vektor ved å velge noen trekk.

Philip = [kaffe, høyde språk]
Philip = [1, 184, 3]
Dette betyr: Philip drikker kaffe, er 184 cm høy, og kan 3 språk.

#### Refleksjoner
Hva lærer man av å studere student vektorene?
- Man kan sammenligne studenter matematisk. For eksempel kan vi finne hvem som er mest like basert på høyde, språk, interesser eller andre trekk.

Hva er forskjellen mellom deg i virkeligheten og deg som vektor?
- En vektor er en forenkling. Den fanger bare noen få egenskaper, ikke hele personen. I virkeligheten har du personlighet, erfaringer, interesser, verdier og mye mer.

Hvilke andre trekk kunne vi ha lagt til for at vektoren ble mer meningsfull?
- alder, studieprogram, programmeringserfaring, treningsvaner, ...
- Jo flere relevante trekk vi legger til, jo mer meningsfull kan sammenligningen bli.

#### Undervisningsrom som Vektorrom
Si vi har et vektorrom med 2 dimensjoner, x-aksen er Høyde og y-aksen er Antall språk.
Hva betyr det når to personer står nær hverandre i dette rommet?
Hvilke trekk forklarer avstanden?
Hvordan kan vi tolke rommet?

##### Svar
Vi har et 2D-rom:
x-akse = høyde
y-akse = antall språk

En person kan være et punkt:
Philip = (184,3)

En annen person kan være:
Ole = (169, 6)

Hva betyr det at to personer står nær hverandre?
Det betyr at de er like på de trekkene vi måler. Her betyr det at de har omtrent samme høyde og kan omtrent like mange språk.

Hvilke trekk forklarer avstanden?
Avstanden forklares bare av:
høydeforskjell + forskjell i antall språk.

Hvordan tolker vi rommet?
Rommet er en forenklet modell av personer. Nærhet i rommet betyr likhet basert på de valgte trekkene.

#### Setningslikhet: Refleksjon
Oppgave:
4. Se på setning 1-6. Hvilke setninger mener du betyr omtrent det samme, og hvilke er forskjellige?

##### Svar
Setninger som ligner:

S1: Mange liker vafler i Norge.
S5: Mange liker ski i Norge.
De er like siden de deler ordene:
mange, liker, i, norge

S2 og S6 ligner også:
S2: Det er mange fjell i Norge.
S6: Det er mange fjell.
De handler om fjell begge to.

S4 er litt annerledes fordi den handler mer om vafler som norsk mat:
S4 = Vafler er populær norsk mat.


#### Setningslikhet: Beregninger
Oppgave:
Ved å bruke vokabularet og Bag-of-Words-representasjonene fra forrige oppgave, se på følgende setninger:
- Mange liker vafler i Norge.
- Det er mange fjell i Norge.
- Vafler er populær norsk mat.

Beregn cosinus-likhet mellom:
- "Mange liker vafler i Norge" og "Det er mange fjell i Norge".
- "Mange liker vafler i Norge" og "Vafler er populær norsk mat".

#### Setningslikhet: Diskusjon
Hvilken av de to setningene: "Det er mange fjell i Norge" og "Vafler er populær norsk mat" er likest "Mange liker vafler i Norge"?
Hva er mulige forklaringer på dette?
Hvordan kunne vi forbedret representasjonen av setningene?

##### Svar
Hvilken setning er likest S1?
S1: Mange liker vafler i Norge.

Sammenligninger:
S1 og S2 = 0.55
S1 og S4 = 0.20

Derfor sier Bag-of-Words og cosinus at:
"Det er mange fjell i Norge" er mest lik "Mange liker vafler i Norge".

Hvorfor?
Fordi S1 og S2 deler flere ord: mange, i, norge

Mens S1 og S4 bare deler: vafler

Men dette føles semantisk rart
S1 handler om: vafler + Norge
S4 hanfler også om: vafler + norsk mat

Så mennesker kan mene at S4 egentlig er mer tematisk lik S1.
Dette viser en svakhet med enkel Bag-of-Words:
Den ser bare på eksakte ord, ikke betydningen.

Hvordan forbedre representasjonen?
Vi kunne brukt:
TF-IDF
slik at vanlige ord som "mange", "i", og "er" får mindre vekt.
Vi kunne også brukt:
stemming/lemmatisering
slik at "Norge" og "norsk" kobles nærmere.
Enda bedre kune vi brukt:
embeddings
som kan forstå at "Norge", "norsk", "vafler" og "mat" kan være semantisk relaterte.

#### Kort oppsummering
Vektorrom-representasjon betyr at vi gjøre personer, ord eller setnigner om til tallvektorer. Da kan vi bruke matematikk til å måle likhet. 
Euklidsk avstand måler hvor langt punkter er fra hverandre.
Cosinuslikhet målet hvor lik retning vektorene har.
For tekst kan vi lage Bag-of-Words-vektorer ved å telle hvilke ord som finnes i hver setning. En svakhet er at modellen bare ser eksakte ord, ikke dypere semantisk mening. Derfor kan TF-IDF, normalisering og embeddings gi bedre representasjoner.
