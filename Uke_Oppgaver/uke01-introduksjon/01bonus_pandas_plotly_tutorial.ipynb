# Uke 1: Bonus - Pandas & Plotly Tutorial
This bonus tutorial introduces two additional libraries commonly used in data science: **Pandas** and **Plotly**.

```python
# Importing libraries:
import pandas as pd
import plotly.express as px
```

If you already know NumPy, Pandas builds on it with labeled `Series` and `DataFrame` objects for tabular data. 
A Series is like a one-dimensional array with labels (similar to a single column in a spreadsheet), while a DataFrame is a two-dimensional table made up of multiple Series sharing the same index.
In practice, most data work in Pandas is done with DataFrames, since they allow easy manipulation, cleaning, and analysis of entire datasets.

```python
# Creating a Pandas DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40],
    "Score": [85, 90, 88, 92]
}
df = pd.DataFrame(data)
print(df)
```

### Reading data
`read_csv` loads data into a `DataFrame`. 

```python
# Reading a CSV file
df = pd.read_csv("01bonus_pandas_plotly_tutorial_data.csv")
# Displaying the first few rows of the DataFrame
print(df.head())
```

Methods like `head`, `tail`, `columns`, and `shape` help you quickly understand the structure of your data.

```python
# Displaying first 10 rows of the DataFrame
print(df.head(10))

# Displaying the last few rows of the DataFrame
print(df.tail())

# Displaying the columns of the DataFrame
print(df.columns)

# Displaying the shape of the DataFrame
print(df.shape)
```

### Common operations
Pandas makes it easy to manipulate tabular data. Here are some common operations:

```python
# DataFrame operations
# Accessing columns
print(df[["Name", "Age"]])

# Filtering rows based on a condition
print(df[df["Age"] > 30])

# Sorting the DataFrame
df_sorted = df.sort_values("Score", ascending=False)
print(df_sorted)
```

### Saving data
Write results back to disk with `to_csv`, `to_parquet`, etc.

```python
# Saving the DataFrame to a CSV file
df_sorted.to_csv("01bonus_pandas_plotly_tutorial_data_sorted.csv", index=False)
```

There are many plotting libraries in Python and Plotly is one of the most popular ones.
Below are some basic examples of how to create different types of plots with Plotly Express, a high-level interface for Plotly.

```python
# Plotting with Plotly
# Basic scatter plot
fig = px.scatter(df, x='Name', y='Score', title='Student Scores by Name (scatter plot)')
fig.show()
```

```python
# Basic bar plot
fig = px.bar(df, x='Name', y='Score', title='Student Scores by Name (bar plot)')
fig.show()
```

```python
# Histogram
fig = px.histogram(df, x='Score', nbins=5, title='Student Scores Distribution (histogram)')
fig.show(renderer="notebook_connected")
```

```python
# Line plot
time_data = {'Week': [1, 2, 3, 4, 5],
             'Score': [75, 80, 82, 85, 90]}
df_time = pd.DataFrame(time_data)

fig = px.line(df_time, x='Week', y='Score', title='Score Trend Over Time', markers=True)
fig.show(renderer="notebook_connected")
```
