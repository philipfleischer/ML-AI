# Conceptual Questions for Chapter 1

### 1. Hva er maskinlæring?
Maskinlæring er når datamaskiner lærer mønstre fra data i stedet for at vi programmerer alle regelene manuelt. Systemet bruker tidligere eksempler til å gjøre prediksjoner eller ta beslutninger om nye tilfeller.

Eksempler som bruker ML:
- Netflix/Youtube - anbefalinger, som lærer hva du liker.
- Spamfilter i e-post, som lærer å kjennne igjen søppelpost.

Eksempler som ikke bruker ML:
- En vanlig kalkulator, som følger faste matematiske regler.
- En enkel alarmklokke, som ringer på et bestemt tidspunkt uten å lære noe.

### 2. Hva er de to hovedgrenene i ML?
De to hovedgrenene innnen Maskin Læring er "veiledet læring" og "ikke-veiledet læring".

I veiledet læring trener vi modellen på data der fasiter er kjent. For eksempel kan modellen lære å skille mellom bilder av katter og hunder fordi treningsbildene er merket som "katt" eller "hund".

I ikke-veiledet læring har vi ikke fasit. Modellen prøver heller å finne møsntre eller grupper i dataen selv. Et eksempel er å gruppere kunder basert på kjøpsvaner uten at vi på forhånd vet hvilke kundegrupper som finnes.

### 3. Hva er copmutational thinking?
Computational thinking betyr å løse problemer på en strukturert måte slik at løsningen kan beskrives som klare steg, gjerne slik en datamaskin kan følge dem.

Et eksempel er å planlegge den raskeste veien til skolen. Man kan dele problemet opp i mindre deler: finne mulige ruter, sammenligne reisetid, velge beste rute og lage en regel for hva man gjør hvis det er forsinkelser. Da bruker man en algoritmisk måte å tenke på.

### 4. To problemer ML har skapt
1. Bias og urettferdighet
ML-modeller lærer fra data. Hvis dataene inneholder skjevheter, kan modellen også ta skjeve beslutninger. For eksempel kan et ansettelsessystem diskriminere dersom treningsdataene historisk favoriserte én gruppe mennesker.

2. Personvernproblemer
ML-systemer trenger ofte store mengder data. Dette kan være problematisk hvis dataene inneholder sensitiv informasjon om personer. For eksempel kan helse-, bilde- eller brukerdata misbrukes dersom de lagres eller behandles dårlig.

### 5a. Kalenderfargin med rød, grønn og blå
Her bruker vi computational thinking ved å først dele opp problemet: Vi må se på hver måned for seg, fordi fargemønsteret starter på nytt hver måned. Deretter finner vi et mønster: fargene går alltid slik: rød, grønn, blp, rød, grønn, blå, osv. Til slutt lager vi en algoritme: For hver måned teller vi hvor mange ganger hver farge brukes, basert på antall dager i måneden.

# Hands-on Problems
## Problem 1.1 - Høyde og vekt. 
Vi har data for høyder fra 61 til 72 inches. Oppgaven spør om vi kan gjette vekten for:
- 60 inches
- 73 inches
- 65.5 inches
- 68.5 inches
Dette er et eksempel på regresjon, fordi vi prøver å furutsi en tallverdi: vekt.

En enkel måte å tenke på:
Når høyden øker, ser vi at vekten stort sett også øker. Derfor kan vi gjette verdier mellom eller litt utenfor tabellen ved å se på mønsteret.

For 65.5 inches (INTERPOLASJON)
65.5 ligger mellom: 
- 65 inches -> 145 lb
- 66 inches -> 151 lb

Da tar vi omtrent midt mellom:
(145 + 151)/2 = 148
Svar: ca. 148 lb

Dette gjør vi likt for 68.5 inches også og får:
Svar: ca. 168.5 lb

For 60 inches   (EKSTRAPOLASJON)
60 er lavere enn den laveste verdien i tabellen.
- 61 inches -> 122 lb
- 62 inches -> 124 lb
Vekten øker med omtrent 2 lb fra 61 til 62, så vi kan gjette at 60 inches er litt lavere enn 122 lb, og kan predikere at verdien for 60lb vil være 120lb.
Svar: ca. 120 lb.

Vi gjør det samme bare motsatt vei for 73 lb.


Kort eksamenssvar:
For å løse oppgaven ser jeg på sammenhengen mellom høyde og vekt. Når høyden øker, øker vekten vanligvis også. For høyder som ligger mellom to kjente verdier, bruker jeg gjennomsnittet av de to nærmeste verdiene. Dette kalles INTERPOLASJON. For høyder utenfor tabelllen, som 60 og 73 inches, forlenger jeg mønsteret litt. Dette kalles EKSTRAPOLASJON. Dette er relatert til regresjon fordi vi bruker kjente data til å furutsi en ukjent numerisk verdi.

## Problem 1.2 - hvilket kylligfor er best?
Her skal vi finne hvilket for som gir størst kyllinger. Da bør vi sammenligne gjennomsnittsvekten for hver fortype.

Gjennomsnittlig vekt per fortype:
- Horsebean - 160 lb
- Linseed - 219 lb
- Soybean - 246 lb
- Sunflower - 329 lb
- Meatmeal - 277 lb
- Casein - 324 lb

Svar:
Sunflower ser ut til å være det beste foret, fordi kyllingene som fikk sunflower hadde høyest gjennomsnittsvekt, omtrent 329 lb.
Casein er nesten like bra, med omternt 324 lb, men sunflower er litt høyere.

Kort eksamenssvar:
For å løse oppgaven, grupeprer jeg kyllingene etter hvilken type for de fikk. Deretter regner jeg ut gjennomsnittsvekten for hver gruppe. Siden oppgaven sier at større kyllinger er bedre, velger jeg foret med høyest gjennomsnittsvekt. Basert på tabellen er sunflower det beste foret, fordi det har høyest gjennomsnittlig vekt.

## Problem 1.3 - Bilforsikring
her skal vi velge forsikringsselskap basert på kundevurderinger. Dette tester enkel datadrevet beslutningstaking: vi grupperer data etter kategori og sammenligner gjennomsnitt.

Vi regner gjennomsnitt for hver av de tre forsikringsselskapene.

GEICO:          (4.7 + 8.3 + 9.2)/3 = 7.4
Progresssive :  (7.4 + 6.7 + 8.9)/3 = 7.67
USAA:           (3.8 + 6.3 + 8.1)/3 = 6.07

Hvis vi hadde sett etter beste mulige enkeltvurdering ville vi ha valgt GEICO, som fikk en rating på 9.2, men den samlede vurdering av de tre personene endrer altså valget vårt.

Svar: Jeg ville valgt Progressive, fordi de har høyest gjennomsnittlig rating, omtrent 7.67 av 10.

Kort eksamenssvar:
For å løse oppgaven, så grupperer jeg ratingene etter forsikringsselskap og regner ut gjennomsnittet for hver gruppe. Dette er en enkel form for mønstergjenkjenning og datadrevet valg. Progressive har høyest gjennomsnittlig rating, så basert på disse dataene virker Progressive som det beste valget. Samtidig er datasetet lite, så konklusjonen er usikker.


## Problem 1.4 - Filmvalg basert på skuespiller
Her skal vi bruke tidligere filmratinger til å forutsi hvilke nye filmer vi sannsynligvis bør se. Dette tester idéen bak enkel prediksjon/anbefaling basert på historiske data.

Regn ut gjenommsnitt per skuespiller.

Irfan Khan: (6.0 + 6.8 + 7.4)/3 = 6.73
Anupam Kher: (7.1 + 2.6)/2 = 4.85
Anil Kapoor: (6.0 + 6.6)/2 = 6.3

Svar: Jeg ville først valgt en ny film med Irfan Khan, fordi filmene hans har høyst avg rating i datasettet.

Kort eksamnssvar:
For å løse oppgaven grupperer jeg filmene etter hovedskuespiller og regner ut gjennomsnittlig IMDB-rating for hver skuespiller. Dette kan brukes som en enkel anbefalingsmodell: hvis en skuespillers tidligere filmer ogte har høy rating, antar vi at en ny film med samme skuespiller også kan være verdt å se. Irfan Khan har høyest gjennomsnitt, omtrent 6.73, og er derfor det beste valget basert på dataene. Datasettet er lite, så dette er bare en enkel prediksjon, ikke en sikker fasit.
