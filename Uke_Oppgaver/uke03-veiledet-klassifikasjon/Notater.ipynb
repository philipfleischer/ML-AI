# Veiledet Klassifikasjon med k-Nærmeste Naboer og evaluering

### Refleksjon og diskusjon
Hva er veiledet læring?

##### Svar:
- Veiledet læring betyr at modellen lærer fra eksempler der fasiten er kjent.
Eksempel:
Input: Bilde av en katt
Fasit/ label y: katt
Modellen får mange slike eksempler og lærer sammenhengen mellom inputdata og riktig svar.

Hva er klassifisering?

##### Svar:
Klassifisering er når modellen skal velge én klasse/kategori.
Eksempel:
E-post -> spam / ikke spam
Bilde -> katt / hund
Pasient -> syk / frisk
Produkt -> elektronikk / klær / møbler
Forskjellen fra regresjon er at klassifisering gir en kategori, mens regresjon gir et tall.

### k-Nærmeste Naboer (kNN)

I løpet av forelesningen lærte dere om kNN-algoritmen, som brukes til klassifisering. Hvordan fungerer algoritmen? Hva er rollen til k?

##### Svar:
kNN klassifiserer et nytt punkt ved å se på de k nærmeste naboene i treningsdataene.
Enkelt sagt:
1. Finn de nærmeste punktene til det nye punktet.
2. Se hvilke klasser naboene har.
3. Klassen som flest naboer har, blir prediksjonen.

Rollen til k er å bestemme hvor mange naboer modellen skap se på.
Lav k-verdi kan være følsom for støy. 
Høy k-verdi gir ofte mer stabil prediksjon, men kan overse lokale mønstre.

Klassifiser den rosa forkanten ved å bruke KNN for k=1, k=3, k=5.

##### Svar
For k=1, så blir det category 1.
For k=3, så blir det category 1.
For k=5, så blir det category 2.


### Scenario: Data splitting
Scenario:
"Tenk deg at du jobber i et stort nettbutikk-selskap. Teamet ditt utvikler en modell som automatisk skal plassere nye produkter i riktig kategori, for eksempel elektronikk, klær, møbler, bøker eller leker.

For hvert produkt har dere informasjon som pris, størrelse, materiale, merke, og en kort produktbeskrivelse. Basert på denne informasjonen skal modellen lære å bestemme hvilken kategori produktet tilhører. Målet er å gjøre opplastnig av produkter raskere og hjelpe kunder med å finne det de leter etter.

Dere har et treningsdatasett med 50 000 produkter, jevnt fordelt på kategoriene (omtrent 10 000 produkter per kategori). For hvert produkt kjenner dere den riktige kategorien. Nå må dere bestemme hvordan dataene skal deles opp for trening og evaluering av modellen."

Spørsmål:
- Ville du trent og testet modellen på de samme dataene? Hvorfor eller hvorfor ikke?
- Hva ville skje hvis du trente og testet en kNN-modell på de samme dataene?
- Hvordan kan du dele dataene slik at evalueringen er rettferdig og representativ for den virkelige oppgaven?

##### Svar:
Ville du trent å testet på samme data?
Nei. Hvis vi trener og tester på samme data, tester vi egentlig bare om modellen husker treningsdataene, ikke om den fungerer på nye produkter.
Målet er å vite om modellen klarer å klassifisere nye, ukjente produkter.

Hva skjer hvis kNN trenes og testes på samme data?
For kNN blir dette spesielt problematisk.
Hvis k=!, vil hvert testpunkt ofte finne seg selv som nærmeste nabo. Da kan modellen få kunstig høy nøyaktighet, kanskje nesten perfekt.
Men dette betyr ikke at modellen faktisk er god på nye data.
Dette kalles ofte overfitting eller overtilpasning.

Hvordan dele data rettferdig?
Et vanlig valg:
80 % treningssett og 20% testsett.

Med 50 000 produkter:
40 000 til trening og 10 000 til testing.

Siden datasettet er balansert, bør testsettet også ha omtrent lik fordeling:
ca. 2 000 produkter fra hver kategori i testsettet.
Dette kalles stratifisert splitting.

### Scenario (endret!): Data splitting
Scenario:
"Tenk deg at du jobber i et stort nettbutikk-selskap. Teamet ditt utvikler en modell som automatisk skal plassere nye produkter i riktig kategori, for eksempel elektronikk, klær, møbler, bøker eller leker.

For hvert produkt har dere informasjon som pris, størrelse, materiale, merke, og en kort produktbeskrivelse. Basert på denne informasjonen skal modellen lære å bestemme hvilken kategori produktet tilhører. Målet er å gjøre opplastnig av produkter raskere og hjelpe kunder med å finne det de leter etter.

Dere har et treningsdatasett med 50 000 produkter. Tenk deg nå at datasettet ikke er balansert: klær har 25 000 produkter, elektronikk 15 000, møbler 6 000, bøker 3 000 og leker 1 000."

Spørsmål:
- Må strategien din for å dele dataene endres? Hvordan?
- Oppsummering: Hvordan vil du definere trenings- og testsettet? Hva er funksjonen deres?
- Noen ganger bruker vi også en tredje type dataset, kalt valideringssett. Hva er formålet med dette?

##### Svar:
Nå er fordelingen:
Klær: 25 000
Elektronikk: 15 000
Møbler: 6 000
Bøker: 3 000
Leker: 1 000

Må strategien endres?
Ja. Vi må være ekstra nøye med at både treningssettet og testsettet får omtrent samme klassefordeling som originaldataene.
Dette kalles fortsatt stratifisert splitting.
Hvis vi ikke gjør det, kan testsettet få for få eksempler av små klasser som "leker".

Eksempel på 80/20 split:
Klær - Totalt 25 000 - Trening 80% 20 000 - Test 20% 5000
Elektronikk - Totalt 15 000 - Trening 80% 12 000 - Test 20% 3000
Møbler - Totalt 6 000 - Trening 80% 4 800 - Test 20% 1200
Bøker - Totalt 3 000 - Trening 80% 2 400 - Test 20% 600
Leker - Totalt 1 000 - Trening 80% 800 - Test 20% 200

Hva er treningssett og testsett?
Treningssett: Data modellen lærer fra.
Testsett: Data modellen ikke har sett før, brukt til å måle hvor godt den generaliserer.

Hva er valideringssett?
Et valideringssett brukes til å velge innstilinger før den endelige testen.
For kNN kan vi bruke valideringssettet til å velge beste k.

Eksempel:
Prøv k = 1, 3, 5, 7 på valideringssettet.
Velg den k-verdien som fungerer best.
Test til slutt én gang på testsettet.

Typisk split:
70 % trening
15 % validering
15 % test


### Forvirringsmatrise 1
![IMG](img/kreft1.png)

##### Svar:
I kreft-eksempeler er positiv klasse:
Kreft

Da betyr:
TP = true positive = faktisk kreft, predikert kreft
TN = true negative = faktisk ingen kreft, predikert ingen kreft
FP = false positive = faktisk ingen kreft, men predikert kreft
FN = false negative = faktisk kreft, men predikert ingen kreft

Viktig:
FP = falsk alarm
FN = sykdom overses

I medisinske tester er FN ogte mest alvorlig, fordi en syk pasient kan bli sendt hjem.


###### Forvirringsmatrise Model 1
Tabell:
                       Predikert kreft     Predikert ingen kreft
Faktisk kreft          16                  4
Faktisk ingen kreft    6                   74

Dermed:
TP = 16
FN = 4
FP = 6
TN = 74

Nøyaktighet / accuracy:
accuracy = (TP + TN) / (TP + TN + FP + FN)
accuracy = (16 + 74) / 100 = 0.90
Accuracy = 90%

Presisjon / precision:
precision = TP / TP + FP
precision = 16 / (16+6) = 0.727
Presisjon = 73%

Betyr: Av dem modellen sa hadde kreft, hvor mange hadde faktisk kreft?

Sensitivitet / recall:
recall = TP / (TP + FN)
recall = 16 / (16 + 4) = 0.80
Sensitivitet = 80%

Betyr: Av alle som faktisk hadde kreft, hvor mange fant modellen?


###### Forvirringsmatrise Model 2
Tabell:
                       Predikert kreft     Predikert ingen kreft
Faktisk kreft          2                   18
Faktisk ingen kreft    0                   80

Dermed:
TP = 2
FN = 18
FP = 0
TN = 80

Nøyaktighet / accuracy:
accuracy = (TP + TN) / (TP + TN + FP + FN)
accuracy = (2 + 80) / 100 = 0.82
Accuracy = 82%

Presisjon / precision:
precision = TP / TP + FP
precision = 2 / (2+0) = 1.0
Presisjon = 100%

Betyr: Når modellen sier kreft, har den alltid rett. men den sier kreft veldig sjelden.

Sensitivitet / recall:
recall = TP / (TP + FN)
recall = 2 / (2 + 18) = 0.10
Sensitivitet = 10%

Betyr: Modellen finner bare 10% av krefttilfellene.

##### Svar:
Sammenligning av modellene

Mål             Model 1     Model 2
Nøyaktighet     90%         82%
Presisjon       73%         100%
Sensitivitet    80%         10%
Falske negative 4           18
Falske positive 6           0

Hvilken modell er bedre?
For kreftdiagnose er Model 1 best.
Grunnen er at Model 1 finner langt flere av de syke pasientene:
Model 1 finner 16 av 20 kreftpasienter.
Model 2 finner bare 2 av 20 kreftpasienter.

Model 2 har høy presisjon, men den overser altfor mange krefttilfeller.

Hvis målet er å aldri overse kreft, da er viktigste verdi:
Sensitivitet / recall

Fordi sensitivitet måler hvor mange av de faktiske syke modellen klarer å finne.

her er Model 1 best:
Model 1: 80%
Model 2: 10%

Hvis målet er å unngå å skremme friske pasienter, da er viktigste verdi:
Presisjon
eller å hå få:
falske positive

Model 2 har:
FP = 0
Presisjon = 100%

Så Model 2 gir ingen falske kreftalarmer, men den er farlig dårlig til å oppdage faktisk kreft.


### Andre klassifiseringsproblemer

Falske negative er mest alvorlige når vi ikke må overse noe farlig.
Eks:
- Krefttest sier frisk, men pasienten har kreft.
- Brannalarm sier ingen brann, men det brenner.
- Svindelsystem sier trygg transaksjon, men det er svindel.
- Sikkerhetssystem slipper inn en farlig person.

Her er problemet at vi ikke oppdager noe farlig.

Falske positive er mest alvorlige når feil alarm har stor kostnad.
Eks:
- Frisk pasient får beskjed om mulig kreft.
- Uskyldig person flagges som kriminell.
- Vanlig bankkort sperres som svindel-
- Normal e-post havner i spam.
- Student mistenkes feilaktig for juks.

Her er problemet a systemet anklager eller stopper noe som egentlig er normalt.

## Kort oppsummering

Veiledet læring betyr at modellen lærer fra data med fasit. Klassifisering betyr å forutsi en klasse, for eksempel kreft eller ingen kreft. kNN klassifiserer et nytt punkt ved å se på de k nærmeste naboene og velge flertallsklassen. Data bør deles i treningssett og testsett, fordi vi vil måle om modellen fungerer på nye data, ikke bare husker treningsdataene. ved ubalanserte datasett bør vi bruke stratifisert splitting slik at klassefordelingen bevares.

I en forvirringsmatrise er TP riktige positive, TN riktige negative, FP falske positive og FN falske negative.
Nøyaktighet måler totalt antall riktige svar, presisjon måler hvor ofte positive prediksjoner er riktige, og sensitivitet måler hvor mange av de faktiske positive tilfellene modellen finner. I medisinske tester er sensitivitet ofte ekstra viktig, fordi falske negaive kan bety at syke pasienter ikke får behandling.
